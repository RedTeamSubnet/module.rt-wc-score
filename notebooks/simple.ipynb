```python
import numpy as np

import matplotlib.pyplot as plt

def scoring_function(count, min_, max_, min_score=0.80, max_score=0.65):
	if count < min_:
		distance_factor = (min_ - count) / min_
		base_score = min_score + (1 - min_score) * distance_factor
		return min(100, base_score + np.cos(count) * 0.0084)
	elif count > max_:
		distance_factor = (count - max_) / max_
		base_score = max_score + (1 - max_score) * distance_factor
		return min(100, base_score + np.cos(count) * 0.0084)
	else:
            return 0


min_ = 1000
max_ = 2000


# Generate test data points
test_counts = [0, 50, 100, 150, 200, 300, 400, 500, 550, 600, 650, 700, 750, 800]
test_counts = [i for i in range(0, 5001,100) if not (1000 <= i <= 2000)]

# Print results
print("\nScoring Results:")
print("-" * 40)
print(f"{'Count':>8} | {'Score':>10}")
print("-" * 40)
for count in test_counts:
    score = scoring_function(count, min_, max_)
    print(f"{count:8d} | {score:10.2f}")
```
